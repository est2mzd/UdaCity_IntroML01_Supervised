```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Lasso
import matplotlib.pyplot as plt
```

```python
# ヘッダ名がないデータなので、列数を指定する
train_data = pd.read_csv("data_quiz_27.csv")
X = train_data.iloc[:, :-1] # 最後の列の1個手前まで
y = train_data.iloc[:,-1]   # 最後の列を指定
```

```python
# インスタンスの作成
lasso_reg = Lasso()

# 最適化
lasso_reg.fit(X,y)
y_pred = lasso_reg.predict(X)

# 係数の取得
reg_coef = lasso_reg.coef_
print(reg_coef)
```

```python
print("X.shape = ", X.shape)
print("type(X) = ", type(X))
print("y.shape = ", y.shape)
print("type(y) = ", type(y))
print("------------------")
print(X.head())
```

##  Xの各列とy / y_pred の関係を見る

```python
# プロット
plt.rcParams['figure.figsize'] = (12,9)
plt.figure()
#
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.scatter(X.iloc[:,i], y,      c='blue');
    plt.scatter(X.iloc[:,i], y_pred, c='red');
    plt.grid(); plt.title( i+1 );
    plt.xlim(-20,20)

print("Blue = Original_Data")
print("Red  = Prediction_Data")
```

## 念のため、LinearRegressionの結果を見る

```python
from sklearn.linear_model import LinearRegression
model_Linear = LinearRegression()
model_Linear.fit(X, y)
y_pred = model_Linear.predict(X)
```
